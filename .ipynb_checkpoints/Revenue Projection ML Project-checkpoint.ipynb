```python
#Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

ledger =  pd.read_csv('UpdatedLedger.csv')
#ledger_later =  ledger[ledger['Day']>=32]
#ledger_later['Net Daily Revenue'] = ledger_later.loc[ledger_later['Details'] == 'Revenue'].groupby('Day')['Amount'].sum()
#ledger_later[ledger_later['Net Daily Revenue'].notnull()]

revenue_rows = ledger.loc[ledger['Details'] == 'Revenue']
revenue_by_day = revenue_rows.groupby('Day')
#transaction_id
net_revenue_by_day = revenue_by_day['Amount'].sum()
#ledger_later['Net Daily Revenue'] = net_revenue_by_day
net_revenue_by_day
```

```python
# Calculate the number of days in the dataset
num_days = len(ledger['Day'].unique())

# Calculate the number of days to include in the training set
num_train_days = round(num_days * 0.8)

# Find the last day to include in the training set
last_train_day = sorted(ledger['Day'].unique())[num_train_days - 1]

#ledger['Revenue'] = ledger.apply(lambda x: x['Amount'] if x['Details'] == 'Revenue', axis=1)
#ledger['Net Daily Revenue'] = ledger.groupby('Day')['Revenue'].sum()
#ledger['Net Daily Revenue'] = ledger.loc[ledger['Details'] == 'Revenue'].groupby('Day')['Amount'].sum()

#ledger_short = ledger.groupby('Day').agg({'Net Daily Revenue': 'last', 'Balance': 'last'}).reset_index()
#ledger_short.columns = ['Day','Net Daily Revenue', 'Closing Balance']
# Split the data into training and testing sets based on the day column
train_data = ledger[ledger['Day'] <= last_train_day]
test_data = ledger[ledger['Day'] > last_train_day]
train_data.to_csv('train_data.csv',index=False)
test_data.to_csv('test_data.csv',index=False)
# Print the number of samples in each set
print(f'Training set size: {len(train_data)}')
print(f'Testing set size: {len(test_data)}')
#ledger_short
#ledger[ledger['Net Daily Revenue'].notnull()]# & (ledger['Details']=='Revenue')]
```
