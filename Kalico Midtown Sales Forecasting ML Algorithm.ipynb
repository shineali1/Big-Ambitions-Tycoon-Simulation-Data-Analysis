```python
'''
Day 548 is a Tuesday.
Doing a few calculations, we find that Day 1 is a Monday. Therefore, we can write a function that would return the 
day of the week based on what remainder that day returns when divided by 7, for example, if day%7=1, it's a Monday, 
if it's 2, then it's a tuesday, so on and so forth.
'''
def day_week(day):
    rem = day % 7
    days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

    if rem == 1:
        return days[0]
    elif rem == 2:
        return days[1]
    elif rem == 3:
        return days[2]
    elif rem == 4:
        return days[3]
    elif rem == 5:
        return days[4]
    elif rem == 6:
        return days[5]
    elif rem == 0:
        return days[6]

day_week(548)

        
  
```

```python
#We write a function day_break to find out if there are any breaks in our data, 
#ie. have we missed data from any period between day 205 and 547
def day_break(day_dump):
    for i in range(len(day_dump)-2):
        break_points = []
        if (day_dump[i+1]-day_dump[i] !=1):
            break_points.append(day_dump[i])
        else:
            continue
        
        return break_points
            

day_break(day_dump)
```

```python
#Random Forest Model performs better than the Decision Tree Model as can be observed from the R^2 value of 0.78.
#The MSE value of 21498.47 means that the average error in predicting the sales value is likely to be 146.6

#LSTM Model

from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

# Split into training and testing sets
train_size = int(len(df) * 0.7)
train_set = df.iloc[:train_size, 1:2].values
test_set = df.iloc[train_size:, 1:2].values

# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
train_set_scaled = scaler.fit_transform(train_set)

# Create input and output sequences for the LSTM model
x_train = []
y_train = []
for i in range(60, len(train_set_scaled)):
    x_train.append(train_set_scaled[i-60:i, 0])
    y_train.append(train_set_scaled[i, 0])
x_train, y_train = np.array(x_train), np.array(y_train)

# Reshape the input data for LSTM model
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))

# Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(LSTM(units=50))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the LSTM model
model.fit(x_train, y_train, epochs=100, batch_size=32)

# Prepare the test data
inputs = df.iloc[len(df) - len(test_set) - 60:, 1:2].values
inputs = scaler.transform(inputs)
x_test = []
for i in range(60, len(inputs)):
    x_test.append(inputs[i-60:i, 0])
x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

# Predict the sales
predicted_sales = scaler.inverse_transform(model.predict(x_test))

# Calculate the R^2 score and MSE
mse = mean_squared_error(test_set, predicted_sales)
r2 = r2_score(test_set, predicted_sales)

print(f"R^2 score: {r2}")
print(f"MSE: {mse}")

# Plot actual vs predicted sales
plt.plot(test_set)
plt.plot(predicted_sales)
plt.legend(['Actual', 'Predicted'])
plt.show()
```

```python
#ARIMA Model

from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

df = kalico_sales_data.drop('Profit',axis=1)
#Arima requires continuous data, therefore we are forced to delete data for days before when break in frequency has occured.
df=df[df['Day']>=373]  #Obtained from day_break function above.
#function to return a date when a day number is inserted. This function assumes the year to start from 2021.
def return_date(day):
    # Convert day number to datetime
    date = pd.Timestamp('2021-01-01') + pd.Timedelta(days=day-1)
    return date

# Add the year to the datetime objects
#Converting the Day column to a datetime object
df['Day'] = df['Day'].apply(lambda x: return_date(x))

df
```

```python
#Step 1: Plot the time series data to see if there is any seasonality, irregularity or trend in the data.
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(10, 6))

# Plot the time series
plt.plot(df['Day'], df['Sales'])

# Set the x-axis label
plt.xlabel('Date')

# Set the y-axis label
plt.ylabel('Sales')

# Show the plot
plt.show()


```

```python
#Step 2: Perform tests to understand if the time series is stationary:
# Perform Augmented Dickey-Fuller test

from statsmodels.tsa.stattools import adfuller

result = adfuller(df["Sales"])
print("ADF statistic:", result[0])
print("p-value:", result[1])
for key, value in result[4].items():
    print("Critical value %s: %.3f" % (key, value))
```

```python
''' 
The ADF statistic is more negative than the critical values at the 1%, 5%, and 10% levels, 
which means we can reject the null hypothesis and conclude that the time series is stationary 
with a high level of confidence. The p-value is also less than 0.05, which further supports the 
conclusion that the time series is stationary.
'''

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# create ACF and PACF plots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,8))
plot_acf(df['Sales'].dropna(), lags=25, ax=ax1)
plot_pacf(df['Sales'].dropna(), lags=25, ax=ax2)
plt.show()
```

```python
'''
Based on the ACF and PACF plots above, we see that at lag 21, the ACF has a significant correlation and 
PACF drops off sharply. This means that there is a seasonality in the data of about 21 days and we will
be using the SARIMA model. The SARIMA model combines the ARIMA model with a seasonal component that models
the periodic behavior of the time series. We specify the seasonal component of the SARIMA model using the 
notation SARIMA(p,d,q)(P,D,Q)m, where (p,d,q) are the non-seasonal ARIMA parameters, (P,D,Q) are the seasonal 
ARIMA parameters, and m is the number of periods in a season.
'''
```

```python
#SARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

# Perform train-test split
train_data = df['Sales'][:-30]
test_data = df['Sales'][-30:]

# Define the SARIMA model
order = (0, 0, 1)
seasonal_order = (0, 1, 1, 21)
model = SARIMAX(train_data, order=order, seasonal_order=seasonal_order)

# Fit the model
model_fit = model.fit()

# Make predictions for the test data
forecast = model_fit.forecast(steps=len(test_data))

# Compute R-squared and RMSE values
r2 = r2_score(test_data, forecast)
rmse = np.sqrt(mean_squared_error(test_data, forecast))

# Print the R-squared and RMSE values
print('R-squared:', r2)
print('RMSE:', rmse)

# Plot the predicted vs actual sales values
plt.figure(figsize=(12,6))
plt.plot(test_data.index, test_data.values, label='Actual Sales')
plt.plot(test_data.index, forecast.values, label='Predicted Sales')
plt.legend()
plt.xlabel('Day')
plt.ylabel('Sales')
plt.title('Actual vs Predicted Sales')
plt.show()

```
